```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
bk = pd.read_csv("Bankruptcy.csv",index_col=0)
bk
```

```python
X = bk.drop(['D', 'YR'], axis=1)
```

```python
X
```

```python
y = bk['D']
```

```python
lor = LogisticRegression(solver = 'saga')
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
```

```python
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

```python
params = {'penalty': ['l1', 'l2', 'elasticnet', None],
          'l1_ratio': np.linspace(0, 1, 5),
         'solver':['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag','saga']}
```

```python
gcv = GridSearchCV(lor,param_grid=params,cv=kfold)
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
pd_cv = pd.DataFrame(gcv.cv_results_)
pd_cv
```

```python
#fit model with best parameters on entire dataset
bm = LogisticRegression(l1_ratio=0.0,penalty='l1',solver='liblinear')
```

```python
bm.fit(X,y)
```
