```python
from sklearn.metrics import r2_score
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.model_selection import KFold
```

```python
bos = pd.read_csv("/Users/darshmac/Documents/cdac/For DBDA/Advance Analytics/Datasets/Boston.csv")
```

```python
X = bos.drop("medv",axis=1)
y = bos["medv"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=23)
```

```python
####
elastic = ElasticNet()
mm_scaler = MinMaxScaler()
knn = KNeighborsRegressor()
pipe_knn = Pipeline([('SCL',mm_scaler),('KNN',knn)])
dtr = DecisionTreeRegressor(random_state=23)
```

```python
## voting
voting = VotingRegressor([('E',elastic),('P_KNN',pipe_knn),('DTR',dtr)])
```

```python
voting.fit(X_train,y_train)
```

```python
y_pred = voting.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
```

```python
############ GRidSearchCV
kfold = KFold(n_splits=5,shuffle=True,random_state=23)
params = {'E__alpha':[0.001,1,2],
          'E__l1_ratio':[0.001,0.5,1],
         'P_KNN__KNN__n_neighbors':[3,5,7],
          'DTR__max_depth':[None,3,6]
         }
```

```python
gcv = GridSearchCV(voting,param_grid=params,cv=kfold,scoring='r2')
```

```python
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```
