>>cat boost works withput hot encoding(without creating dummies)

```python
from xgboost import XGBClassifier
```

```python
import pandas as pd
import numpy as np
```

```python
hr = pd.read_csv("/Users/darshmac/Documents/cdac/For DBDA/PML/Cases/human-resources-analytics/HR_comma_sep.csv")
hr
```

```python
dum_hr = pd.get_dummies(hr,drop_first=True)
X = dum_hr.drop('left',axis=1)
y = dum_hr['left']
```

```python
from sklearn.model_selection import StratifiedKFold,GridSearchCV,train_test_split
```

```python
kflod = StratifiedKFold(n_splits=5, shuffle=True,random_state=23)
```

```python
xgbc = XGBClassifier(random_state=23)
```

```python
xgparam = {'n_estimators':[25,50,75,100]}
xgbgcv = GridSearchCV(xgbc,param_grid=xgparam,cv=kflod,verbose=3,scoring='neg_log_loss')
```

```python
%%time
xgbgcv.fit(X,y)
```

```python
print(xgbgcv.best_params_)
print(xgbgcv.best_score_)
```

```python
#For Boston
bos = pd.read_csv("/Users/darshmac/Documents/cdac/For DBDA/PML/Cases/Boston Housing/Boston.csv")
bos
```

```python
X = bos.drop('medv', axis =1)
y = bos['medv']
```

```python
from xgboost import XGBRegressor
xgbr = XGBRegressor(random_state=23)
```

```python
from sklearn.model_selection import KFold
rkfold = KFold(n_splits=5,shuffle=True,random_state=23)
```

```python
xgbrparam = {'n_estimators':[25,50,75,100],
             'learning_rate':np.linspace(0.001,0.7,5),
             'max_depth':[None,2,3,5]}
xgbrgcv = GridSearchCV(xgbr,param_grid=xgbrparam,cv=rkfold)
```

```python
%%time
xgbrgcv.fit(X,y)
```

```python
print(xgbrgcv.best_params_)
print(xgbrgcv.best_score_)
```
