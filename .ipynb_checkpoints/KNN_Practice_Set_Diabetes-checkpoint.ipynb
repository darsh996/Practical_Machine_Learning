```python
X = nzbk.drop('Personal.Loan',axis=1)
```

```python
y = nzbk['Personal.Loan']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=23,shuffle=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import accuracy_score, log_loss
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
knn = KNeighborsClassifier(n_neighbors =1)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
y_pred_prob = knn.predict_proba(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(log_loss(y_test,y_pred_prob))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True,random_state=23)
```

```python
print(knn.get_params())
```

```python
params = {'n_neighbors':np.arange(1,20,1)}
```

```python
gcv = GridSearchCV(knn,param_grid=params,
                  cv = kfold,
                  scoring='accuracy')
```

```python
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
y_pred_prob = knn.predict_proba(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(log_loss(y_test,y_pred_prob))
```

```python
#pipeline
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
std_scaler = StandardScaler()
```

```python
#pipe for std scaler
pipe_std = Pipeline([('SCL',std_scaler),('KNN',knn)])
```

```python
pipe_std.fit(X_train,y_train)
```

```python
y_pred = pipe_std.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
#for minmaxscaler
minmaxscl = MinMaxScaler()
```

```python
pipmin = Pipeline([('MXSCL',minmaxscl),('KNN',knn)])
```

```python
pipmin.fit(X_train,y_train)
```

```python
y_pred = pipmin.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
```
