```python
bk = pd.read_csv("Bankruptcy.csv",index_col=0)
bk
```

```python
X = bk.drop(['D','YR'],axis=1)
y = bk['D']
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
params = {'solver':['svd', 'lsqr', 'eigen']}
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gcv = GridSearchCV(lda,param_grid=params,cv = kfold,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
print(gcv.best_score_)
print(gcv.best_params_)
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
qda = QuadraticDiscriminantAnalysis()
```

```python
params = {'reg_param': np.linspace(0,1,20)}
```

```python
qgcv = GridSearchCV(qda,param_grid=params,cv = kfold,scoring='neg_log_loss')
```

```python
qgcv.fit(X,y)
```

```python
print(qgcv.best_score_)
print(qgcv.best_params_)
```
