```python
sat = pd.read_csv('Satellite.csv',delimiter=';')
sat
```

```python
X = sat.drop('classes',axis=1)
y = sat['classes']
```

```python
y.nunique()
```

```python
y.value_counts()
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
params = {'solver':['svd', 'lsqr', 'eigen']}
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
lgcv = GridSearchCV(lda,param_grid=params,cv = kfold,scoring='neg_log_loss')
```

```python
lgcv.fit(X,y)
```

```python
print(lgcv.best_score_)
print(lgcv.best_params_)
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

```python
qda = QuadraticDiscriminantAnalysis()
```

```python
params = {'reg_param': np.linspace(0,1,20)}
```

```python
qgcv = GridSearchCV(qda,param_grid=params,cv = kfold,scoring='neg_log_loss')
```

```python
qgcv.fit(X,y)
```

```python
print(qgcv.best_score_)
print(qgcv.best_params_)
```

```python
#best model
b_model = lgcv.best_estimator_
```

```python
tst_sat = pd.read_csv("tst_satellite.csv")
tst_sat
```

```python
predictions = b_model.predict(tst_sat)
```
