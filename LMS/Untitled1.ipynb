```python
X = dum_tel.drop('Churn_Yes',axis=1)
y = dum_tel['Churn_Yes']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=23)#,stratify=y)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
from sklearn.svm import SVC
#svm = SVC(kernel='linear',probability=True,random_state=23)
svm = SVC(kernel='linear')
#svm = SVC(kernel='rbf')
svm.fit(X_train,y_train)
```

```python
y_pred = svm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23) 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'C':np.linspace(0.0001,10,20)}
#for radial rbf
#params = {'C':np.linspace(0.0001,10,20),
#        'gamma':np.linspace(0.001,5,10)}
```

```python
gcv = GridSearchCV(svm,param_grid=params,cv = kfold)#,scoring='accuracy')#scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
print(gcv.best_score_)
print(gcv.best_params_)
```

```python
pd_cv = pd.DataFrame(gcv.cv_results_)
pd_cv
```
