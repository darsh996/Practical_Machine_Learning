```python
hr1 = pd.read_csv("HR_comma_sep.csv")
hr = pd.get_dummies(hr1,drop_first=True)
hr
```

```python
X = hr.drop(['left'],axis=1)
X
```

```python
y = hr['left']
y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.3,random_state=23)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
from sklearn.ensemble import BaggingClassifier

bagg = BaggingClassifier(lr,n_estimators=15)
bagg.fit(X_train,y_train)
```

```python
y_pred_prob = bagg.predict_proba(X_test)
```

```python
from sklearn.metrics import log_loss
print(log_loss(y_test,y_pred_prob))
```

```python
#gcv
from sklearn.model_selection import GridSearchCV,StratifiedKFold

kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

```python
bagg.get_params()
```

```python
bagg = BaggingClassifier(lr,n_estimators=15,random_state=23)
```

```python
params = {'estimator__penalty':['l1','l2','elasticnet',None]}
```

```python
gcv = GridSearchCV(bagg,param_grid=params,cv=kfold,scoring='neg_log_loss')
```

```python
%%time
#start_time = time.time()
gcv.fit(X,y)
#print(time.time()-start_time)
```

```python
print(gcv.best_params_)
print(gcv.best_score_)
```
