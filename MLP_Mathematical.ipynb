```python
#Import Required Libraries
import matplotlib.pyplot as plt
import numpy as np

#Dataset name is Salary_Dataset
years_experience = np.array([1.2, 1.4, 1.6])
salary = np.array([39344, 46206, 37732])

# Initial weights and bias
w = 0.5
b = 0.2

# Learning rate
learning_rate = 0.01

# Number of iterations
epochs = 100

# Training loop
losses = []
for epoch in range(epochs):
    # Forward propagation
    predictions = w * years_experience + b

    # Compute loss
    loss = np.mean((predictions - salary)**2)
    losses.append(loss)

    # Backpropagation (updating weights and bias)
    w = w - learning_rate * np.mean((predictions - salary) * years_experience)
    b = b - learning_rate * np.mean(predictions - salary)

# Visualize the training process
plt.plot(range(epochs), losses, label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Over Time')
plt.legend()
plt.show()
```

```python
#Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Name of Dataset is Salary_Dataset
years_experience = np.array([1.2, 1.4, 1.6])
salary = np.array([39344, 46206, 37732])

# Loss function
def loss_function(w, b):
    predictions = w * years_experience + b
    loss = np.mean((predictions - salary)**2)
    return loss

# Generate a grid of parameter values
w_values = np.linspace(-1, 1, 100)
b_values = np.linspace(-50000, 50000, 100)
W, B = np.meshgrid(w_values, b_values)
loss_values = np.zeros_like(W)

# Calculate loss for each parameter combination
for i in range(len(w_values)):
    for j in range(len(b_values)):
        loss_values[i, j] = loss_function(W[i, j], B[i, j])

# Find the global minima
min_loss_idx = np.unravel_index(np.argmin(loss_values), loss_values.shape)
min_w = W[min_loss_idx]
min_b = B[min_loss_idx]
min_loss = np.min(loss_values)

# Plot the 3D surface
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W, B, loss_values, cmap='viridis', alpha=0.8)
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('Loss')
ax.set_title('3D Loss Surface')

# Plot the minimum point
ax.scatter(min_w, min_b, min_loss, color='red', s=100, label='Global Minima')

# Plot the starting point
ax.scatter(0.5, 0, loss_function(0.5, 0), color='blue', s=100, label='Initial Point')

# Annotate the global minima values
ax.text(min_w, min_b, min_loss, f'Global Minima\n(w={min_w:.2f}, b={min_b:.2f}, Loss={min_loss:.2f})',
        fontsize=10, color='blue', va='bottom', ha='center')

ax.legend()
plt.show()
```

```python
#Import Required Libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the loss function (replace this with your actual loss function)
def loss_function(x, y):
    return x**2 + y**2 + np.sin(5*x) + np.cos(5*y)

# Define the gradient of the loss function
def gradient(x, y):
    dx = 2*x + 5*np.cos(5*x)
    dy = 2*y - 5*np.sin(5*y)
    return np.array([dx, dy])

# Gradient descent optimization
def gradient_descent(starting_point, learning_rate, iterations):
    current_point = starting_point
    trajectory = [current_point]

    for _ in range(iterations):
        grad = gradient(*current_point)
        current_point = current_point - learning_rate * grad
        trajectory.append(current_point)

    return np.array(trajectory)

# Generate data points for the loss function
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
x, y = np.meshgrid(x, y)
z = loss_function(x, y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the loss function surface
ax.plot_surface(x, y, z, cmap='viridis')

# Add labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Loss')

# Mark the global minimum
min_point = np.unravel_index(np.argmin(z), z.shape)
ax.scatter(x[min_point], y[min_point], z[min_point], color='red', marker='o', label='Global Minimum')

# Run gradient descent from an initial point
initial_point = np.array([2.0, -2.0])  # You may need to adjust the initial point based on your function
learning_rate = 0.1
iterations = 50
trajectory = gradient_descent(initial_point, learning_rate, iterations)

# Plot the trajectory of gradient descent
ax.plot(trajectory[:, 0], trajectory[:, 1], loss_function(trajectory[:, 0], trajectory[:, 1]),
        color='blue', marker='o', linestyle='dashed', label='Gradient Descent')

# Show the plot
plt.legend()
plt.show()
```
