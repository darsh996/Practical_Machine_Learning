```python
from lightgbm import LGBMClassifier
```

```python
import pandas as pd
import numpy as np
```

```python
#hr = pd.read_csv("/Users/darshmac/Documents/cdac/For DBDA/PML/Cases/human-resources-analytics/HR_comma_sep.csv")
hr = pd.read_csv("/kaggle/input/hr-comma-sep/HR_comma_sep.csv")
hr
```

```python
dum_hr = pd.get_dummies(hr,drop_first=True)
X = dum_hr.drop('left',axis=1)
y = dum_hr['left']
```

```python
from sklearn.model_selection import StratifiedKFold,GridSearchCV,train_test_split
```

```python
kflod = StratifiedKFold(n_splits=5, shuffle=True,random_state=23)
```

```python
lgbc = LGBMClassifier(random_state=23)
```

```python
lgparam = {'n_estimators':[25,50,75,100],
             'learning_rate':np.linspace(0.001,0.7,5),
             'max_depth':[None,2,3,5]}
lgbgcv = GridSearchCV(lgbc,param_grid=lgparam,cv=kflod,verbose=3,scoring='neg_log_loss')
```

```python
%%time
lgbgcv.fit(X,y)
```

```python
print(lgbgcv.best_params_)
print(lgbgcv.best_score_)
```

```python
#For Boston
#bos = pd.read_csv("/Users/darshmac/Documents/cdac/For DBDA/PML/Cases/Boston Housing/Boston.csv")
bos = pd.read_csv("/kaggle/input/boston/Boston.csv")
bos
```

```python
X = bos.drop('medv', axis =1)
y = bos['medv']
```

```python
from lightgbm import LGBMRegressor
lgbr = LGBMRegressor(random_state=23)
```

```python
from sklearn.model_selection import KFold
rkfold = KFold(n_splits=5,shuffle=True,random_state=23)
```

```python
xlgbrparam = {'n_estimators':[25,50,75,100],
             'learning_rate':np.linspace(0.001,0.7,5),
             'max_depth':[None,2,3,5]}
xlgbrgcv = GridSearchCV(lgbr,param_grid=xlgbrparam,cv=rkfold)
```

```python
%%time
xlgbrgcv.fit(X,y)
```

```python
print(xlgbrgcv.best_params_)
print(xlgbrgcv.best_score_)
```
